```python
def animate(i):
    x.append(xs[i])
    y.append(ys[i])
    s.append(ss[i]/1000)
    sc.set_offsets(np.c_[x,y])
    sc.set_sizes(s)
    ax.legend(grf_data['Country'] + " " + x[i] +" " + size_ax + " " + str(s[i]))
```

```python
for c in set(full_df['Country']):
    x_ax = 'Year'
    y_ax = 'Population'
    size_ax = 'capacity_mw'
    
    grf_data = full_df[full_df['Country']==c].reset_index()

    fig, ax = plt.subplots()
    x,y,s  = [],[],[]
    sc = ax.scatter(x,y,s=s)

    xs = grf_data[x_ax]
    ys = grf_data[y_ax]
    ss = grf_data[size_ax]
    plt.xlim(xs.min(),xs.max())#set to min and max
    plt.ylim(ys.min(),ys.max())#set to min and max
    ani = FuncAnimation(fig, animate, interval=100, repeat=True) 
    try:
        ani.save('output/'+c+ ' '+y_ax+'.gif', writer='pillow')
    except:
        pass
```
