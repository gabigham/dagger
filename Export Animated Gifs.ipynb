```python
%matplotlib notebook
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
matplotlib.use('Agg')
from merge_data import merge_data
import numpy as np
import pandas as pd
import os
from config import (census_key, gkey)
import gmaps
```

```python
#os.chdir('C://Users//adam//Desktop//dagger')

cap_df, con_df = merge_data()
con_df = con_df.sort_values(by='Year')
cap_df['per_cap_cap'] = cap_df['cap_added']/cap_df['Population']
```

```python
cap_df.head()
```

```python
def animate(i):
    x.append(xs[i])
    y.append(ys[i])
    s.append(ss[i])
    sc.set_offsets(np.c_[x,y])
    sc.set_sizes(s)
    ax.legend(grf_data['Country'] + " " + x[i] +" " + size_ax + " " + str(round(s[i],2)))
```

```python

x_ax = 'Year'
y_ax = 'HDI'
size_ax = 'per_cap_cap'

for c in set(cap_df['Country']):
    grf_data = cap_df[cap_df['Country']==c].reset_index()
    fig, ax = plt.subplots()
    x,y,s  = [],[],[]
    sc = ax.scatter(x,y,s=s)

    xs = grf_data[x_ax]
    ys = grf_data[y_ax]
    ss = grf_data[size_ax]
    plt.xlim(int(xs.min()),int(xs.max()))#set to min and max
    plt.ylim(ys.min(),ys.max())#set to min and max
    ani = FuncAnimation(fig, animate, interval=250, repeat=True) 
    try:
        ani.save('output/'+c+ ' '+y_ax+'.gif', writer='pillow')
    except:
        pass
    plt.clf()
    plt.cla()
```
